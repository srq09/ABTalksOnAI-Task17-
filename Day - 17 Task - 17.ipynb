```python
# Day 17 - Probability Distribution for Word Occurrences using SciPy

# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import poisson, norm
from collections import Counter
import re

# Sample NLP text dataset
text_data = """
Artificial Intelligence is transforming the world.
AI is used in healthcare, finance, education, and automation.
Probability helps AI systems make intelligent decisions under uncertainty.
Natural Language Processing uses probability distributions for word occurrence analysis.
AI and NLP are core domains in intelligent automation systems.
"""

# Step 1: Text Preprocessing
clean_text = re.sub(r'[^a-zA-Z\s]', '', text_data.lower())
words = clean_text.split()

print("Processed Words:")
print(words)

# Step 2: Count Word Occurrences
word_counts = Counter(words)

print("\nWord Frequency Counts:")
for word, count in word_counts.items():
    print(f"{word}: {count}")

# Convert to DataFrame
df = pd.DataFrame(word_counts.items(), columns=['Word', 'Frequency'])
print("\nWord Frequency DataFrame:")
print(df)

# Step 3: Probability Distribution Calculation
total_words = sum(word_counts.values())
df['Probability'] = df['Frequency'] / total_words

print("\nProbability Distribution:")
print(df.sort_values(by='Probability', ascending=False))

# Step 4: Poisson Distribution Analysis
mean_frequency = np.mean(df['Frequency'])

df['Poisson_Probability'] = poisson.pmf(df['Frequency'], mean_frequency)

print("\nPoisson Distribution Probabilities:")
print(df[['Word', 'Frequency', 'Poisson_Probability']])

# Step 5: Normal Distribution Analysis
mean = np.mean(df['Frequency'])
std = np.std(df['Frequency'])

df['Normal_Probability'] = norm.pdf(df['Frequency'], mean, std)

print("\nNormal Distribution Probabilities:")
print(df[['Word', 'Frequency', 'Normal_Probability']])

# Step 6: Visualization - Word Frequency
plt.figure(figsize=(12, 6))
plt.bar(df['Word'], df['Frequency'])
plt.xticks(rotation=90)
plt.title('Word Frequency Distribution')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# Step 7: Visualization - Probability Distribution
plt.figure(figsize=(12, 6))
plt.bar(df['Word'], df['Probability'])
plt.xticks(rotation=90)
plt.title('Word Occurrence Probability Distribution')
plt.xlabel('Words')
plt.ylabel('Probability')
plt.tight_layout()
plt.show()

# Step 8: Interpretation
print("\nInsights:")
print("1. Higher frequency words have higher occurrence probability.")
print("2. Poisson distribution helps model discrete word occurrence events.")
print("3. Normal distribution helps analyze overall frequency spread.")
print("4. Probability distributions help NLP systems understand uncertainty in language data.")
```

```python
# ABTalks Season-3 Day 17
# Probability Distribution for Word Occurrences using SciPy

# Import Libraries
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
from scipy.stats import poisson

# Sample NLP text corpus
text = """
Artificial intelligence is transforming industries.
AI helps in healthcare, finance, education, and automation.
Natural language processing uses probability to understand word patterns.
AI systems rely on statistical reasoning to handle uncertainty.
Probability distributions help model word occurrences in NLP tasks.
"""

# Step 1: Text Cleaning
clean_text = re.sub(r'[^a-zA-Z\s]', '', text.lower())

# Step 2: Tokenization
words = clean_text.split()

print("Processed Words:")
print(words)

# Step 3: Count Word Frequencies
word_counts = Counter(words)

print("\nWord Frequency Counts:")
for word, count in word_counts.items():
    print(f"{word}: {count}")

# Step 4: Create DataFrame
df = pd.DataFrame(word_counts.items(), columns=["Word", "Frequency"])

# Step 5: Empirical Probability Distribution
total_words = df["Frequency"].sum()
df["Probability"] = df["Frequency"] / total_words

print("\nProbability Distribution:")
print(df.sort_values(by="Probability", ascending=False))

# Step 6: Poisson Distribution (SciPy)
lambda_value = df["Frequency"].mean()
df["Poisson_Probability"] = poisson.pmf(df["Frequency"], lambda_value)

print("\nPoisson Probability Distribution:")
print(df[["Word", "Frequency", "Poisson_Probability"]])

# Step 7: Visualization - Frequency Distribution
plt.figure(figsize=(12, 6))
plt.bar(df["Word"], df["Frequency"])
plt.xticks(rotation=90)
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.title("Word Frequency Distribution")
plt.tight_layout()
plt.show()

# Step 8: Visualization - Probability Distribution
plt.figure(figsize=(12, 6))
plt.bar(df["Word"], df["Probability"])
plt.xticks(rotation=90)
plt.xlabel("Words")
plt.ylabel("Probability")
plt.title("Word Occurrence Probability Distribution")
plt.tight_layout()
plt.show()

# Step 9: Insights
print("\nInsights:")
print("1. Words with higher frequency have higher probability of occurrence.")
print("2. Poisson distribution models discrete word occurrence events.")
print("3. Probability helps NLP systems understand uncertainty in text data.")
print("4. This demonstrates probabilistic thinking in Natural Language Processing.")
```
